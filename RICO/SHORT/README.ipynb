# RICO composite case - SHORT version - 16-19 December 2004

## Description

The case referred to as **RICO/SHORT** ...

## Case setup

The case setup follows the description at http://projects.knmi.nl/rico/setup1d_composite.html. It is close to the LES setup of van Zanten et al. (2011), except that the surface flux formulation was imposed and the LESs only last one day.

### General specifications

The case starts on 16 December 2004, 00:00 UTC, lasts 72 hours and finish on 19 December 2004, 00:00 UT. The latitude is 18°N and the longitude is 61.5°W. The surface pressure is set to 1015.4 hPa and is supposed to remain constant along the case.

### Boundary conditions

The surface temperature is prescribed and held constant to $t_s = 299.8$ K.

### Lateral forcing

## SCM-enabled file

#### Hypotheses so far

  * The vertical grid has 10-m resolution for all fields, from the surface to 60000 m. The time axis goes from 00:00 UTC, 16 December 2004 to 00:00 UTC, 19 December 2004 with a 1-hour timestep.
  
  * Pressure is computed using the altitude, the temperature and surface pressure, assuming a dry atmosphere;
  * $q_t$ is assumed to be equal to $q_v$;
  * $r_t$ is computed from $q_t$ and $r_v$ is assumed to be equal to $r_t$;
  * $\left.\dfrac{\partial \theta}{\partial t}\right)_{\text{LS}}$ is computed from $\left.\dfrac{\partial T}{\partial t}\right)_{\text{LS}}$ using the initial pressure levels;
  * We assume $\left.\dfrac{\partial q_t}{\partial t}\right)_{\text{LS}} = \left.\dfrac{\partial q_v}{\partial t}\right)_{\text{LS}}$;
  * We assume $\left.\dfrac{\partial r_t}{\partial t}\right)_{\text{LS}} = \dfrac{q_t}{1-q_t}\left.\dfrac{\partial q_t}{\partial t}\right)_{\text{LS}}$ using the initial profile of $q_t=q_v$;
  * We assume $\left.\dfrac{\partial r_v}{\partial t}\right)_{\text{LS}} = \left.\dfrac{\partial r_t}{\partial t}\right)_{\text{LS}}$.
