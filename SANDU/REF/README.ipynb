# SANDU composite case - REF version - 15-18 July 2006

## Description

The case referred to as **SANDU/REF** ...

## Case setup

The case setup is described in Sandu and Stevens (2011). It was also used in the LES intercomparison of de Roode et al. (2016). The full documentation of the case is available [here](https://raw.githubusercontent.com/GdR-DEPHY/DEPHY-SCM/master/SANDU/description_composite.pdf), with an addendum [here](https://raw.githubusercontent.com/GdR-DEPHY/DEPHY-SCM/master/SANDU/description_composite_addendum.txt) (relaxation of wind, temperature and humidity). Besides we use the input file composite_ref_init.nc downloaded at http://gop.meteo.uni-koeln.de/~neggers/transitions/.

### General specifications

The case starts on 15 July 2006<sup id="a1">[1](#f1)</sup>, 18:00 UTC (10:00 LT), lasts 72 hours and finish on 18 July 2006, 18:00 UTC. The latitude is 25°N and the longitude is 125°W. The surface pressure is set to 1016.811 hPa and is supposed to remain constant along the case.

### Initial State

The case is originally defined in terms of liquid potential temperature $\theta_l$ and total water content $q_t$. The initial profiles are idealized between the surface and 3200 m:

\begin{equation}
\theta_l(z) = \begin{cases}
 290.97 & \text{if } 0 \leq z \leq 902.52 \text{ m}\\
 302.13 + 5.7 \times 10^{-3} \times (z-945.54) & \text{if } 945.54 \leq z \leq 3200 \text{ m}
\end{cases}
\end{equation}

\begin{equation}
q_t(z) = \begin{cases}
 10.46 & \text{if } 0 \leq z \leq 902.52 \text{ m}\\
 4.12 - 0.37 \times 10^{-3} \times (z-945.54) & \text{if } 945.54 \leq z \leq 3200 \text{ m}
\end{cases}
\end{equation}

Linear interpolation is used at the inversion between 902.52 and 945.54 m.
The wind reads for $0 \leq z \leq 3200$ m:
\begin{eqnarray}
u(z) & = & -2.25\\
v(z) & = & -5.69+1.44\times 10^{-3} \times z
\end{eqnarray}

The profiles were then vertically extended with ERA-Interim up to 47973.81 m.

### Boundary conditions

The surface temperature is prescribed and evolves in time:

Surface fluxes are computed by the model and assume a parameterization of $z_0$.

### Large scale forcing

No large-scale advection of temperature or humidity is prescribed, but A relaxation is applied to the temperature and humidity above 3000 m, towards their initial profiles (see addendum). The relaxation timescale is set to 1 hour<sup id="a2">[1](#f2)</sup>.

A vertical velocity $w$ is also prescribed and is to be used to compute the vertical advection of all model state variables $u$, $v$, $T$, $q_v$, $q_l$...
It is set to $w=-1.86 \times 10^{-6} \times z$ for $0 \leq z \leq 2000$m then constant above 2000 m and progressively comes back to the ERA-Interim profile.

A large-scale horizontal pressure gradient is applied through the forcing of a constant geostrophic wind $(u_g,v_g)$ which is equal to the initial wind profile $(u,v)$.

### Microphysics

The cloud droplet concentration should be imposed to 100 cm$^{-3}$.

## SCM-enabled file

#### Hypotheses so far

  * The vertical grid has 10-m resolution for all fields, from the surface to 47970 m. The time axis goes from 18:00 UTC, 15 July 2006 to 18:00 UTC, 15 July 2006 with a 30-minute timestep. All conversion assume no liquid water ($q_l=0$, so that $q_v=q_t$ and $r_v=r_t$. This is definitely something to improve in the future.
  
  * Pressure is computed using the altitude, the liquid potential temperature ($\theta=\theta_l$ and surface pressure, assuming a dry atmosphere;
  * $q_t$ is assumed to be equal to $q_v$ (no liquid water);
  * $r_t$ is computed from $q_t$ and $r_v$ is assumed to be equal to $r_t$ (also no liquid water);
