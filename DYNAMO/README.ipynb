# CINDY2011/DYNAMO cases

## Description

The CINDY2011/DYNAMO (Cooperative Indian Ocean Experiment on Intraseasonal Variability in Year 2011‐DYNAmic of the Madden‐Julian Oscillation) field campaign was conducted over the tropical Indian Ocean from October to December 2011 and aims at improving our knowledge on processes at play during the initiation of the MJO in that region (Yoneyama et al., 2013). One specific objective of the campaign was to document the moisture and energy budgets over two arrays (approximately 800 × 800 km each) and their intraseasonal variability: one was located north of the Equator (Northern Sounding Array - NSA) and one south of the Equator (Southern Sounding Array -SSA). The domains of the campaign were mostly constrained by the locations of suitable facilities on islands and atolls (Johnson and Ciesielski, 2013). 

The period sampled by the CINDY2011/DYNAMO campaign experienced distinct convective regimes in the northern and southern arrays (Johnson and Ciesielski, 2013; Johnson et al., 2015). Over SSA, convection was not well organized and mainly modulated by the InterTropical Convergence Zone (ITCZ) activity. In contrast, three enhanced rainfall events passed over NSA. The first and second active phases (16 October to 3 November and 18 November to 1 December) were classified as MJO events (Achuthavarier et al., 2016; Yoneyama et al., 2013). The third one (10–25 December) has a less coherent propagation and is not characteristic of MJO disturbances.

## Case setup

The case setup follows Abdel-Lathif et al. (2018). Version 3a of array-averaged data is obtained from http://johnson.atmos.colostate.edu/dynamo/products/array_averages/. It is provided on a 25-hPa vertical grid every 3 hours. Because it is based only on array observations, it is less accurate when the Research Vessels were offsite (after 7 December 2011 for NSA, after 30 November 2011 for SSA). Data is supplemented above 50 hPa with ERA-Interim data (Dee et al., 2011) and above 1 hPa with the standard atmosphere of (COESA, 1976).

The NSA and SSA arrays have the following coordinates (cf. http://johnson.atmos.colostate.edu/dynamo/products/array_averages/):

The case covers the 3 month period from 00:00 UTC, 1 October to 21:00 UTC, 31 December 2011. Initial profiles (temperature, specific humidity, zonal, and meridional wind) and forcings are from the CSU version 3a CINDY2011/DYNAMO data set (http://johnson.atmos.colostate.edu/dynamo/products/array_averages/). The atmospheric column is forced (every 3 hour) by:
* horizontal advection of temperature;
* horizontal advection of moisture;
* large‐scale vertical pressure velocity (the model thus computes its own vertical advection of energy and moisture);
* Nudging of the horizontal wind components over the whole column, toward the array-averaged vertical profile and with a 3-h time scale;
* Nudging of the temperature and specific humidity above 50 hPa (upper limit of the CSU data set), toward the array-averaged vertical profile and with a 3-h time scale. This avoids low drift in the upper atmosphere;
* Surface temperature from the Woods Hole Oceanographic Institution (WHOI) averaged over the array by CSU.

The CSU data set has an upper limit at 50 hPa (included). Above, we follow Abdel-Lathif et al. (2018):
* ERA‐Interim reanalysis (Dee et al., 2011) data, averaged over 0-6°N, 73°E-80°E for NSA, over ?? for SSA, is used to supplement the vertical profiles of altitude, pressure, temperature, moisture, horizontal wind and vertical velocity between 50 hPa (excluded) and 1 hPa (included).
* The 1976 U.S. standard atmosphere profile (COESA, 1976) is used to supplement the vertical profile of altitude, pressure and temperature above 1 hPa (excluded), up to 86 km.
* All other fields are set to zero if not mentionned above (eg horizonal advections above 50 hPa, specific humidity above 1 hPa).

For NSA, we use a latitude of 2.5°N and a longitude of 77.3°E. For SSA, we use a latitude of ?? and a longitude of ??. The altitude above the geoide is 0 m. The surface pressure follows the time varying array-averaged observations.

Quicklooks of the initial state and the case forcing are available [here](https://github.com/romainroehrig/DEPHY-SCM/tree/master/DYNAMO/NSA3A/images/driver_DEF) for NSA, and [here](https://github.com/romainroehrig/DEPHY-SCM/tree/master/DYNAMO/SSA3A/images/driver_DEF) for SSA.

### Notes

* The precision in CSU ASCII files (and in the derived CSU netCDF file) is not sufficient for water vapor mixing ratio, especially at upper levels of the troposphere. Therefore, specific humidity is derived from the relative humidity (wrt ice or liquid water depending on temperature), based on the Goff-Gratch equation.
