# FIRE Reference - 14-15 July 1987

## Description

The case referred to as **FIRE/REF** was developed along the European Project on Cloud Systems in Climate Models (EUROCS) to study the diurnal cycle of stratocumulus. The case is based on observations collected in July 1987 during the First International Satellite Cloud Climatology Project (ISCCP) Regional Experiment (FIRE), off the coast of California, more specifically during the FIRE-I observing period (see Duynkerke et al. 2004 and reference therein for details).

## Case setup

The case setup follows the description in Dyunkerke et al. (2004) and Chlond et al. (2004), except that the parameterization of cloud radiative effect was imposed.

### General specifications

The case starts on 14 July 1987, 08:00 UTC (00:00 LT), lasts 37 hours and finish on 15 July 1987, 21:00 UTC. The latitude is 33.3°N and the longitude is 119.5°W. The surface pressure is set to 1012.5 hPa and is supposed to remain constant along the case. Compared to the **FIRE/REF** profile this case has the large-scale forcing =0 at 1200m to be sure that at the top of the model (1200m at least for the LES) the forcing are nul

### Initial State

The case is originally defined in terms of liquid-water potential temperature $\theta_l$ and total water content $q_t$. The inversion layer is near 600 m, where a jump in $\theta_l$ of 12 K and a jump in $q_t$ of $-3$ g kg$^{-1}$ occur. The initial profiles are defined up to 12000 m and a 10-m-deep layer for representing the jump is introduced (Chlond et al. 2004):

\begin{equation}
\theta_l(z) = \begin{cases}
 287.5 & \text{if } 0 \leq z \leq 595 \text{ m}\\
 297.5 + 12\times\dfrac{z-595}{605-595} & \text{if } 595 \leq z \leq 605 \text{ m}\\
 299.5 + 0.0075\times(z-605) & \text{if } 605 \leq z \leq 3200 \text{ m}
\end{cases}
\end{equation}

\begin{equation}
q_t(z) = \begin{cases}
 9.6 & \text{if } 0 \leq z \leq 595 \text{ m}\\
 0.6 - 3\times\dfrac{z-595}{605-595} & \text{if } 595 \leq z \leq 605 \text{ m}\\
 6.6 + 0.003\times(z-605) & \text{if } 605 \leq z \leq 3200 \text{ m}
\end{cases}
\end{equation}

The wind profil is initialized from an estimate of the geostrophic wind, and made constant between the surface and 1200 m : $(u, v) = (3.4, -4.9)$ m s$^{-1}$.

### Boundary conditions

The surface temperature is prescribed and held constant to $t_s = 289$ K.

### Lateral forcing

Large-scale advections of liquid-water potential temperature $\left(\partial_t \theta_l\right)_{\text{LS}}$ and of total water content $\left(\partial_t q_t\right)_{\text{LS}}$ are prescribed and held constant in time. They are defined between the surface and 1200m:

\begin{equation}
\left(\frac{\partial \theta_l}{\partial t}\right)_{\text{LS}} = -7.5\times 10^{-8} \times \max(z,500) \text{ in K s}^{-1}
\end{equation}

\begin{equation}
\left(\frac{\partial q_t}{\partial t}\right)_{\text{LS}} = 3.0\times 10^{-11} \times \max(z,500)  \text{ in kg kg s}^{-1}
\end{equation}

A vertical velocity $w$ is also prescribed and is to be used to compute the vertical advection of all model state variables $u$, $v$, $T$, $q_v$, $q_l$...

\begin{equation}
w = -1.0 \times 10^{-5}\times z \text{ in m s}^{-1}
\end{equation}

Linear interpolation should be used to compute the forcings at each model level.

A large-scale horizontal pressure gradient is applied through the forcing of a constant geostrophic wind $(u_g,v_g)$ which is equal to the initial wind profile $(u,v)$.  

## SCM-enabled file

#### Hypotheses so far

  * The vertical grid has 10-m resolution for all fields between the surface and 2000 m and a 100-m resolution above, up to 40 km. The time axis goes from 08:00 UTC, 14 July 1987 to 21:00 UTC, 15 July 1987 with a 1-hour timestep.
  
  * Pressure is computed using the altitude, the liquid-water potential temperature (assuming it equal to the potential temperature) and surface pressure, considereing water vapor;
  * $q_v$ is assumed to be equal to $q_t$ (thus $q_l = 0$ g kg$^{-1}$);
  * $r_t$ is computed from $q_t$ and $r_v$ is assumed to be equal to $r_t$;
  * $\left(\partial_t T\right)_{\text{LS}}$ is computed from $\left(\partial_t \theta\right)_{\text{LS}}$ using the initial pressure levels and assuming $\left(\partial_t \theta\right)_{\text{LS}} = \left(\partial_t \theta_l\right)_{\text{LS}}$;
  * We assume $\left(\partial_t q_v\right)_{\text{LS}} = \left(\partial_t q_t\right)_{\text{LS}}$;
  * We assume $\left(\partial_t r_t\right)_{\text{LS}} = \dfrac{q_t}{1-q_t}\left(\partial_t q_t\right)_{\text{LS}}$ using the initial profile of $q_t=q_v$;
  * We assume $\left(\partial_t r_v\right)_{\text{LS}} = \left(\partial_t r_t\right)_{\text{LS}}$.
  * Profiles are extended up to 40 km:
    * $(u,v)$ and $(u_g,v_g)$ are simply extended upward, supposing them constant;
    * $q_t$ is linearly extended to 0 from 1200 m to 1300 m, and set to 0 g kg$^{-1}$ above;
    * $\theta_l$ is extended up to 2500 m following the same vertical gradient as below. Above, the ERA5 temperature profile is used, up to 40 km.
    * $\left(\partial_t \theta_l\right)_{\text{LS}}$, $\left(\partial_t q_t\right)_{\text{LS}}$ and $w$ are linearly extended to 0 from 1200 m to 1300 m, and set to 0 above.
